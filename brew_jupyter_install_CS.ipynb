## - 요세미티 쓰시는 분들은 신경 안쓰셔도 될 문제 같습니다.

### 1. Homebrew를 통한 python 설치 및 환경 설정

#### 1) Homebrew 설치
> $ ruby -e "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/master/install)"

#### 2) brew 진단
> $ brew doctor

#### 3) python 설치 
> $ brew install python

#### 4) python 경로 설정
 - python은 기본적으로 mac에서  /usr/bin에 설치되어 있음
 - brew로 설치한 파이썬은 /usr/local/bin에 설치되어 있음
 - PATH 설정을 바꿔주어야 brew를 통해 설치한 python으로 작업 가능함 (굳이 /usr/bin/python을 지울 필요는 없음).
 - 초기 python 경로는 /usr/bin/python

---

#### 5) virtualenv 설치 

> $ pip install virtualenv

#### 6) virtualenv로 가상환경 만들기 
(자신이 가상환경 만들고 싶은 폴더에서 작업하시면 됩니다. 저는 Study폴더 밑에 pyenv라는 폴더를 만들고 그안에 env1이라는 가상 환경을 만들었습니다.) 

> $ mkdir Study/pyenv

> $ cd Study/pyenv

> $ virtualenv --no-site-packages —distribute env1

#### 7) 가상 환경 실행 

> $ source env1/bin/activate

#### 8) pip 로 패키지 설치

> (env1)$ pip install ipython[all] 

> (env1)$ pip install numpy scipy matplotlib

> (env1)$ pip install jsonschema -> ipython3.0 부터 설치 필요

## 결론
- brew 설치가 나름 장점이 있으나 그냥 표준 설치가 더 편할 수도 있음. 
  물론 본인은 아직 안해봐서 모름. 해보고 싶지만 포맷을 당하는 데미지를 입어 더 해보기가 귀찮음.
- 매버릭스 및 그 하위 OS버젼을 사용하고 계신분들께 유용할 것 같음. 
  요세미티는 문제 없다는 여러 증인들의 증언이 있었음. 
  현재 문제 생겼던 사람은 매버릭스 및 스노우레오파드 사용자였음.
- 이 문제를 해결하기까지 많은 도움을 주신 이성호님, 김덕태님, 김무성님께 무한한 감사를 드립니다.

## 추가 팁
- Jupyter에서 R을이용해 install.package() 사용시 cran 주소를 적지 않으면 설치가 안됨. install.packages(“패키지 이름”, repos =“cran주소”) 이렇게 해야 설치가 됨.
- 매번 cran 주소를 치기 귀찮으므로 홈폴더에 .Rprofile 이란 파일을 만들어 cran 주소 지정

> $ vi .Rprofile

local({

    r <- getOption("repos")

    r["CRAN"] <- "http://cran.nexr.com/"

    options(repos = r)

})

:wq!

- 본인은 한국 대표 cran인 nexr 주소를 사용함. 본인이 맘에 드는 cran 주소를 사용하면 됨
- 각 국가의 cran 주소는 아래 홈페이지에 있음

> http://cran.r-project.org/mirrors.html

- 그리고 다시 접속하여 패키지를 설치하면 install.packages(“패키지명”)으로 설치가 가능함
